- Alice in Wone

```python
with open('data/Alice.txt') as file:
    text = file.read()
```

```python
STOPWORDS # 불용어  # nltk에 있는 stopwords 
```

```python
# stopwords
stopwords = STOPWORDS
len(stopwords)
```

```python
#stopwords에 'said' 추가하기
stopwords.add('said') #  set 으로 한 이유는 중복을 베제하기 위해서 set으로 함 
```

```python
wc = WordCloud(background_color='white',max_words=2000,stopwords=stopwords)
wc = wc.generate(text) # text를 생성하다. 
```

```python
keys = list(wc.words_.keys())
values = list(wc.words_.values())
for i in range(10):
    print(f'{keys[i]}:{values[i]}')
```

```python
type(wc.words_.values)
type(wc.words_.keys)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc,interpolation='bilinear') # 
plt.axis('off')
plt.show()
```

```python
# 그림 위에 표시하기
mask = np.array(Image.open('data/Alice_mask.png'))
plt.figure(figsize=(8,8))
plt.imshow(mask,cmap=plt.cm.gray,interpolation='bilinear') # 
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white',max_words=2000,mask=mask,stopwords=stopwords)
wc = wc.generate(text) # text를 생성하다. 
plt.figure(figsize=(12,12))
plt.imshow(wc,interpolation='bilinear') # 
plt.axis('off')
plt.show()
```

- Star Wars - A New Hope

```python
import numpy as np
with open('data/A_new_hope.txt') as file:
    text = file.read()
text = text.replace('HAN','Han').replace("LUKE's",'Luck')
mask = np.array(Image.open('data/Stormtrooper_mask.png'))
```

```python
stopwords = STOPWORDS
stopwords.add('int')
stopwords.add('ext')
```

```python
wc = WordCloud(background_color=None,max_words=1000,mask=mask,stopwords=stopwords,
                margin=10,random_state=1).generate(text) # text를 생성하다.
default_colors = wc.to_array()
```

```python
# HSL(Hue,Saturation, Lightness) 코드로 변환
def grey_color(word,font_size,position,orientation,random_state=None,**kwargs):
    return f'hsl(0,0%, {np.random.randint(60,101)}%)'
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wc.recolor(color_func=grey_color,random_state=3),interpolation='bilinear') # 
plt.axis('off')
plt.show()
```
